### OUTLIERS

Creamos una función para arreglar los outliers

```python
def arreglarOutliers(data):
    outliers = data.quantile([0.05,0.95])
    n = len(data)
    outliers = list(outliers)
    mean = data.mean()
    median = data.median()
    
    for element in range(n):
        if int(data[element]) < int(outliers[0]):
            data.loc[element]= mean
        elif int(data[element]) > int(outliers[1]):
            data.loc[element]= median

    return data
```

### 3. ¿Qué categorías de videos tienen la mejor proporción (ratio) de “Me gusta” / “No me gusta”? 


```python
temp3 = video_level.copy()

temp3.dislikes = [1 if i==0 else i for i in temp3.dislikes]
temp3['ratio'] = temp3['likes']/temp3['dislikes']
max_ratio_like_dislike = temp3['ratio'].max() #Ver cual es la mejor proporción

ans3 = temp3[temp3['ratio']==max_ratio_like_dislike]
print(f'Número de videos con el mismo ratio:\t{len(ans3)}')
print(f'Las categorías con mejor proporción de “Me gusta” / “No me gusta” son:\n{ans3.category_description.unique()}')

#visualización
g=sns.barplot(x='ratio', y='category_description',data=temp3)

plt.rcParams["figure.figsize"] = (20,20)
plt.xticks(rotation=90)
plt.show()
```

### 9. ¿Los videos en tendencia son los que mayor cantidad de comentarios positivos reciben?

```python
temp9 = video_level.copy()
above_mean = temp9['likes']>temp9['likes'].mean()
below_mean = [not i for i in above_mean]

print('Porcentaje de videos cuyos "Me gusta" superan el promedio:\n',\
    sum(temp9[above_mean].freq)/(sum(temp9[below_mean].freq)+sum(temp9[above_mean].freq))*100, '%')

#visualización
plt.rcParams["figure.figsize"] = (5,5)
plt.pie([sum(temp9[above_mean].freq), sum(temp9[below_mean].freq)],
    labels=['Encima del promedio', 'Debajo del promedio'])
plt.show()
```

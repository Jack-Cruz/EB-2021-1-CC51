### OUTLIERS

Creamos una función para arreglar los outliers

```python
def arreglarOutliers(data):
    outliers = data.quantile([0.05,0.95])
    n = len(data)
    outliers = list(outliers)
    mean = data.mean()
    median = data.median()
    
    for element in range(n):
        if int(data[element]) < int(outliers[0]):
            data.loc[element]= mean
        elif int(data[element]) > int(outliers[1]):
            data.loc[element]= median

    return data
```
